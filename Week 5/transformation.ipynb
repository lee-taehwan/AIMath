```python
import numpy as np
import matplotlib.pyplot as plt

x, y = np.meshgrid(np.arange(-5, 6), np.arange(-5, 6))
print(x) ; print(y)
```

```python
plt.scatter(x, y)
plt.scatter(0, 0, color="orange")
vec1 = np.array([2, 3])
x1, y1 = vec1
plt.arrow(0, 0, x1, y1, width=0.2, length_includes_head=True)
plt.grid()
plt.show()
```

```python
M = np.array([[3, 1], [0, 2]])
print(M)
print(M.shape)
```

```python
t_x, t_y = np.dot(M, np.array([x.flatten(), y.flatten()]))
plt.scatter(t_x, t_y)
plt.scatter(0, 0, color='orange')
vec1 = np.array([2, 3])
t_x, t_y = vec1
plt.arrow(0, 0, t_x, t_y, width=0.2, length_includes_head=True)
plt.grid()
plt.show()
```

```python
M = np.array([[1, 5], [0, 1]])
t_x, t_y = np.dot(M, np.array([x.flatten(), y.flatten()]))
plt.scatter(t_x, t_y)
plt.scatter(0, 0, color='orange')
plt.grid()
plt.show()
```

```python
M = np.array([[2, -2], [1, -1]])
t_x, t_y = np.dot(M, np.array([x.flatten(), y.flatten()]))
plt.scatter(t_x, t_y)
plt.scatter(0, 0, color='orange')
plt.grid()
plt.show()
```
